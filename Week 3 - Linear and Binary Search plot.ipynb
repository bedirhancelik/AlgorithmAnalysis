```python
import random
def random_list(n,min,max):
    list1=[]
    for i in range(n):
        list1.append(random.randint(min,max))
    return list1
random_list(10,1,10)
```

```python
def linear_search(my_list,item_search):
    found=(-1,-1) 
    n=len(my_list)
    for i in range(n):
        if my_list[i]==item_search:
            found=(my_list[i],i)

    return found
```

```python
def my_experimental_study(iterNum=50):
    cost=[]
    x_low,x_high,array_size=-100,100,40
    print("Array size:"+str(array_size))
    for iter in range(iterNum):
        my_list=random_list(array_size,x_low,x_high)
        my_search_item=random_list(1,x_low,x_high)
        my_search_item=my_search_item[0]
        
        result=linear_search(my_list,my_search_item)
        if(result[1]==-1):
            cost.append(array_size)
        else:
            cost.append(result[1])
        print(result)
    return cost
```

```python
c_s_l=my_experimental_study()
```

```python
import matplotlib.pyplot as plt
plt.plot(c_s_l)
```

```python
def my_binary_search(my_list, item_search):
    found=(-1,-1,0)
    low = 0
    high = len(my_list) - 1
    s=0
    while low <= high:
        mid = (low + high) // 2
        print(low,high,mid)
        s=s+1
        if my_list[mid] == item_search:
            return my_list[mid],mid,s
        elif my_list[mid] > item_search:
            high = mid - 1
        else:
            low = mid + 1
    print(s)
    return found[0],found[1],s
```

```python
def my_bubble_sort(my_list):
    n=len(my_list)
    for i in range(n-1,-1,-1):
        for j in range(0,i):
            if not(my_list[j]<my_list[j+1]):
                temp=my_list[j]
                my_list[j]=my_list[j+1]
                my_list[j+1]=temp
    return my_list
```

```python
x_low,x_high,array_size=-10,10,4
my_list=random_list(array_size,x_low,x_high)
my_list=my_bubble_sort(my_list)
print(my_list)
my_search_item=random_list(1,x_low,x_high)
my_search_item=my_search_item[0]
```

```python
def my_experimental_study_binary(iterNum=50):
    const=[]
    x_low=-100
    x_high=100
    array_size=40
    print("array size :",array_size)
    for iter in range(iterNum):
        my_list=random_list(array_size,x_low,x_high) 
        my_list=my_bubble_sort(my_list)
        
        my_search_item=random_list(1,x_low,x_high)
        my_search_item= my_search_item[0]
        
        result=my_binary_search(my_list,my_search_item)
        const.append(result[2])
    return const
```
